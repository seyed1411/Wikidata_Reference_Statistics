```python
###  Plotting triples per nodes distribution

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

dfgene2016 = pd.read_csv('../Query Results/Distribution of Triple per Nodes/genewiki2016.csv', index_col=None, header=0)
dfgene2016['year']='2016'
dfgene2016['subset']='Gene Wiki'
dfgene2021 = pd.read_csv('../Query Results/Distribution of Triple per Nodes/genewiki2021.csv', index_col=None, header=0)
dfgene2021['year']='2021'
dfgene2021['subset']='Gene Wiki'

dftax2016 = pd.read_csv('../Query Results/Distribution of Triple per Nodes/taxonomy2016.csv', index_col=None, header=0)
dftax2016['year']='2016'
dftax2016['subset']='Taxonomy'
dftax2021 = pd.read_csv('../Query Results/Distribution of Triple per Nodes/taxonomy2021.csv', index_col=None, header=0)
dftax2021['year']='2021'
dftax2021['subset']='Taxonomy'

dfast2016 = pd.read_csv('../Query Results/Distribution of Triple per Nodes/astronomy2016.csv', index_col=None, header=0)
dfast2016['year']='2016'
dfast2016['subset']='Astronomy'
dfast2021 = pd.read_csv('../Query Results/Distribution of Triple per Nodes/astronomy2021.csv', index_col=None, header=0)
dfast2021['year']='2021'
dfast2021['subset']='Astronomy'

dfmusic2016 = pd.read_csv('../Query Results/Distribution of Triple per Nodes/music_2016.csv', index_col=None, header=0)
dfmusic2016['year']='2016'
dfmusic2016['subset']='Music'
dfmusic2021 = pd.read_csv('../Query Results/Distribution of Triple per Nodes/music_2021.csv', index_col=None, header=0)
dfmusic2021['year']='2021'
dfmusic2021['subset']='Music'

dflaw2016 = pd.read_csv('../Query Results/Distribution of Triple per Nodes/law_2016.csv', index_col=None, header=0)
dflaw2016['year']='2016'
dflaw2016['subset']='Law'
dflaw2021 = pd.read_csv('../Query Results/Distribution of Triple per Nodes/law2021.csv', index_col=None, header=0)
dflaw2021['year']='2021'
dflaw2021['subset']='Law'

dfships2016 = pd.read_csv('../Query Results/Distribution of Triple per Nodes/ships_2016.csv', index_col=None, header=0)
dfships2016['year']='2016'
dfships2016['subset']='Ships'
dfships2021 = pd.read_csv('../Query Results/Distribution of Triple per Nodes/ships_2021.csv', index_col=None, header=0)
dfships2021['year']='2021'
dfships2021['subset']='Ships'

dfs = [dfgene2016,dfgene2021,dftax2016, dftax2021, dfast2016, dfast2021, dfmusic2016, dfmusic2021, dflaw2016, dflaw2021, dfships2016, dfships2021]
frame = pd.concat(dfs, axis=0, ignore_index=True)

plt.figure(figsize=(12,5))
box_plot = sns.boxplot(x="subset", y="num",
            hue="year", palette=["#3498db", "#2ecc71"],
            data=frame,showmeans=True,showfliers=False,
            meanprops={"marker":"^",
            "markerfacecolor":"black", 
            "markeredgecolor":"black",
            "markersize":"5"},medianprops={'color':'red'},flierprops={"marker":"o", "markersize":"5"})
box_plot.set_xlabel("Subset")
box_plot.set_ylabel("Number of Triples in Each Reference Node")
sns.despine(offset=15, trim=True)
plt.savefig('triples_per_ref_distribution.eps' , format='eps')
```

```python
###  Plotting reference sharing distribution

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

dfgene2016 = pd.read_csv('../Query Results/Distribution of Reference Sharing/genewiki2016.csv', index_col=None, header=0)
dfgene2016['year']='2016'
dfgene2016['subset']='Gene Wiki'
dfgene2021 = pd.read_csv('../Query Results/Distribution of Reference Sharing/genewiki2021.csv', index_col=None, header=0)
dfgene2021['year']='2021'
dfgene2021['subset']='Gene Wiki'

dftax2016 = pd.read_csv('../Query Results/Distribution of Reference Sharing/taxonomy2016.csv', index_col=None, header=0)
dftax2016['year']='2016'
dftax2016['subset']='Taxonomy'
dftax2021 = pd.read_csv('../Query Results/Distribution of Reference Sharing/taxonomy2021.csv', index_col=None, header=0)
dftax2021['year']='2021'
dftax2021['subset']='Taxonomy'

dfast2016 = pd.read_csv('../Query Results/Distribution of Reference Sharing/astronomy2016.csv', index_col=None, header=0)
dfast2016['year']='2016'
dfast2016['subset']='Astronomy'
dfast2021 = pd.read_csv('../Query Results/Distribution of Reference Sharing/astronomy2021.csv', index_col=None, header=0)
dfast2021['year']='2021'
dfast2021['subset']='Astronomy'

dfmusic2016 = pd.read_csv('../Query Results/Distribution of Reference Sharing/music2016.csv', index_col=None, header=0)
dfmusic2016['year']='2016'
dfmusic2016['subset']='Music'
dfmusic2021 = pd.read_csv('../Query Results/Distribution of Reference Sharing/music2021.csv', index_col=None, header=0)
dfmusic2021['year']='2021'
dfmusic2021['subset']='Music'

dflaw2016 = pd.read_csv('../Query Results/Distribution of Reference Sharing/law2016.csv', index_col=None, header=0)
dflaw2016['year']='2016'
dflaw2016['subset']='Law'
dflaw2021 = pd.read_csv('../Query Results/Distribution of Reference Sharing/law2021.csv', index_col=None, header=0)
dflaw2021['year']='2021'
dflaw2021['subset']='Law'

dfships2016 = pd.read_csv('../Query Results/Distribution of Reference Sharing/ships2016.csv', index_col=None, header=0)
dfships2016['year']='2016'
dfships2016['subset']='Ships'
dfships2021 = pd.read_csv('../Query Results/Distribution of Reference Sharing/ships2021.csv', index_col=None, header=0)
dfships2021['year']='2021'
dfships2021['subset']='Ships'

MaxMeans = {'Gene Wiki':[round(dfgene2021['num'].mean()),dfgene2021['num'].max()], 'Taxonomy':[round(dftax2021['num'].mean()),dftax2021['num'].max()],
            'Astronomy':[round(dfast2021['num'].mean()),dfast2021['num'].max()], 'Law':[round(dflaw2021['num'].mean()),dflaw2021['num'].max()],
            'Music':[round(dfmusic2021['num'].mean()),dfmusic2021['num'].max()], 'Ships':[round(dfships2021['num'].mean()),dfships2021['num'].max()]}

dfMaxMeans = pd.DataFrame(MaxMeans, index=['Mean', 'Max'])
dfMaxMeans.to_csv('../ISWC_CSV_data/ref_sharing/mean-max.csv')
dfs = [dfgene2016,dfgene2021,dftax2016, dftax2021, dfast2016, dfast2021, dfmusic2016, dfmusic2021, dflaw2016, dflaw2021, dfships2016, dfships2021]
frame = pd.concat(dfs, axis=0, ignore_index=True)
plt.figure(figsize=(12,5))
box_plot = sns.boxplot(x="subset", y="num",
            hue="year", palette=["#3498db", "#2ecc71"],
            data=frame,showmeans=False,showfliers=False,
            meanprops={"marker":"^",
            "markerfacecolor":"black", 
            "markeredgecolor":"black",
            "markersize":"5"},medianprops={'color':'red'},flierprops={"marker":"o", "markersize":"5"})
box_plot.set_xlabel("Subset")
box_plot.set_ylabel("Number of Statements Pointed to Each Refrence Node (>1)")
sns.despine(offset=15, trim=True)
plt.savefig('statement_connected_to_ref_distribution.eps' , format='eps')
```
